```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as la
import matplotlib.animation as animation
%matplotlib widget
```

```python
fig, ax = plt.subplots(4, 4, figsize=(13,8), dpi=100)
plt.gcf().set_facecolor("white") # set white background

t = np.load("../data/run.npz")["t"]
cov = np.load("../data/run.npz")["cov"]

for i in range(4):
    for j in range(4):
        ax[i, j].tick_params(which='both', bottom=False, labelbottom=False)
        ax[i, j].axhline(0, c='gray', lw=1, ls="dashed")
        ax[i, j].set_yticks([0])
        ax[i, j].set_yticklabels(["0"])

ax[0, 0].plot(t, cov[:, 0])
ax[1, 1].plot(t, cov[:, 1])
ax[0, 1].plot(t, cov[:, 2])
ax[2, 2].plot(t, cov[:, 3])
ax[3, 3].plot(t, cov[:, 4])
ax[2, 3].plot(t, cov[:, 5])
ax[0, 3].plot(t, cov[:, 6])
ax[1, 2].plot(t, cov[:, 7])
ax[0, 2].plot(t, cov[:, 8])
ax[1, 3].plot(t, cov[:, 9])

plt.show()
plt.tight_layout()
```
