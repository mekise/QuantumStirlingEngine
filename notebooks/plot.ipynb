```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as la
import matplotlib.animation as animation
%matplotlib widget
```

```python
filenpz = "../data/run_04.npz"

t = np.load(filenpz)["t"]
cov = np.load(filenpz)["cov"]

# list(np.load(filenpz).keys())
```

```python
print("Th =", np.load(filenpz)["Th"])
print("Tc =", np.load(filenpz)["Tc"])
print("gammah =", np.load(filenpz)["gammah"])
print("gammac =", np.load(filenpz)["gammac"])
print("omegam =", np.load(filenpz)["omegam"])
print("G =", np.load(filenpz)["G"])
print("tend =", np.load(filenpz)["tend"])
```

```python
fig, ax = plt.subplots(4, 4, figsize=(13,8), dpi=100)
plt.gcf().set_facecolor("white") # set white background

for i in range(4):
    for j in range(4):
        ax[i, j].tick_params(which='both', bottom=False, labelbottom=False)
        ax[i, j].axhline(0, c='gray', lw=1, ls="dashed")
        ax[i, j].set_yticks([0])
        ax[i, j].set_yticklabels(["0"])

ax[0, 0].plot(t, cov[:, 0])
ax[1, 1].plot(t, cov[:, 1])
ax[0, 1].plot(t, cov[:, 2])
ax[2, 2].plot(t, cov[:, 3])
ax[3, 3].plot(t, cov[:, 4])
ax[2, 3].plot(t, cov[:, 5])
ax[0, 3].plot(t, cov[:, 6])
ax[1, 2].plot(t, cov[:, 7])
ax[0, 2].plot(t, cov[:, 8])
ax[1, 3].plot(t, cov[:, 9])

plt.show()
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2, dpi=80, figsize=(10,4))
plt.gcf().set_facecolor("white") # set white background

file_eig = "../data/eig_scan.npz"

tspan = np.load(file_eig)["tspan"]
deltap = np.load(file_eig)["deltap"]
omegam = np.load(file_eig)["omegam"]
eigplus = np.load(file_eig)["eigplus"]
eigminus = np.load(file_eig)["eigminus"]

ax[0].axhline(omegam, c='C1')
ax[0].plot(tspan, deltap)

ax[1].plot(tspan, eigplus)
ax[1].plot(tspan, eigminus)

plt.show()
plt.tight_layout()
```
